```python
#Problem Statement
'''You are the data scientist at a retail company ,and the marketing team is looking to optimize thier
promotional campaigns by targetting specific coustemer segments . The goal is to identify distinct groups of coustemer
based on thier demographic information such as age,income and spnding score.'''
#Solution:Agglomerative Hierarchical Clustering
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
```

```python
# Creating the data set
np.random.seed(42)
df=pd.DataFrame({'Age':np.random.randint(18,65,100),'Income':np.random.randint(30000,100000,100),
                 'SpendingScore':np.random.randint(1,100,100)})
print(df)
```

```python
scaler=StandardScaler()
scaler_data=scaler.fit_transform(df)
```

```python
clustering=AgglomerativeClustering(n_clusters=3,metric='euclidean',linkage='ward')
df['Cluster']=clustering.fit_predict(scaler_data)
```

```python
plt.scatter(df['Income'],df['SpendingScore'],c=df['Cluster'],cmap='viridis')
plt.xlabel('Income')
plt.ylabel('SpendingScore')
plt.title('Consumer Segmentation')
plt.show()
```

```python
new_customer=pd.DataFrame({'Age':[30],'Income':[50000],
                 'SpendingScore':[70]})
scaled_new_customer=scaler.transform(new_customer)
predicted_customer=clustering.fit_predict(scaled_new_customer)
print("the new coustemer predicted cluster is :",predicted_customer)
```
