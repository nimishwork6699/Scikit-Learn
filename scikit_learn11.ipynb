```python
#Problem statement
'''Imagine you aare working for a ecomerce company ,and marketing team want to target coustemermore effectively my creating segment
based on their purchasing behaviour.However traditional method are not providing satisfactory answer'''
#Solution : DBSCAN to identify cluster
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
np.random.seed(42)
data=np.random.rand(100,2)
```

```python
scaler=StandardScaler()
data_scaled=scaler.fit_transform(data)
```

```python
dbscan=DBSCAN(eps=0.3,min_samples=5)
clusters=dbscan.fit_predict(data_scaled)
```

```python
plt.scatter(data[:,0],data[:,1],c=clusters,cmap='viridis',marker='o',s=50)
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
result_df=pd.DataFrame({'Feature 1':data[:,0],'Feature 2':data[:,1],'Cluster':clusters})
```

```python
print("Number of clusters:",len(np.unique(clusters)))
print("Size of cluster:")
print(result_df['Cluster'].value_counts())
```

```python
#input
user_input=np.array([[0.6,0.7]])
user_input_scaled=scaler.transform(user_input)
user_cluster=dbscan.fit_predict(user_input_scaled)
print("User belong to cluster:",user_cluster[0])
```
