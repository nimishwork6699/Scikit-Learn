```python
#Problem Statement
'''As a retail buisness owner,you want to understand and categorize your coustemer into distinct segment baased on thier purchasing
behaviour.This segment will help you with tailoring marketing strategies and promotion to group of peple'''
# Solution using gaussian mixture model(GMM)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
```

```python
np.random.seed(42)
segment1=np.random.normal(loc=30,scale=5,size=100)
segment2=np.random.normal(loc=60,scale=10,size=150)
segment3=np.random.normal(loc=90,scale=8,size=120)
data=np.concatenate([segment1,segment2,segment3]).reshape(-1,1)
```

```python
scaler=StandardScaler()
scaler_data=scaler.fit_transform(data)
```

```python
n_components=3
gmm=GaussianMixture(n_components=n_components,random_state=42)
gmm.fit(scaler_data)
```

```python
cluster_label=gmm.predict(scaler_data)
```

```python
plt.scatter(data,np.zeros_like(data),c=cluster_label,cmap='viridis')
plt.xlabel('Purchase Amount')
plt.title('Cluster segmentation')
plt.show()
```

```python
user_input=float(input("Enter a purchase amount to predict the coustemer segment  "))
user_input_scaled=scaler.transform(np.array([[user_input]]))
predict_label=gmm.predict(user_input_scaled.reshape(-1,1))[0]
print("the predicted coustemer segment is :",predict_label+1)
```
