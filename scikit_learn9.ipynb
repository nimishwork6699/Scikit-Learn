```python
# Problem Statement
'''A mall wanted to undrstand its coustemer base better to improve its marketing strategy.The goal is to
cluster the coustemer on the basis of their anual income and spending score,so that marketing team can tailor their
campaign to cluster's preference behaviour '''
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
np.random.seed(42)
data={'AnnualIncome':np.random.randint(30000,100000,100),'SpendingScore':np.random.randint(1,100,100)}
df=pd.DataFrame(data)
print(df)
```

```python
plt.scatter(df['AnnualIncome'],df['SpendingScore'])
plt.title('Coustemer Data')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
X=df.values
kmeans=KMeans(n_clusters=3,random_state=42)
df['Cluster']=kmeans.fit_predict(X)
```

```python
print(df)
```

```python
plt.scatter(df['AnnualIncome'],df['SpendingScore'],c=df['Cluster'],cmap='rainbow')
plt.title('KMean Clustering')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
user_input={'AnnualIncome':[75000],'SpendingScore':[70]}
user_df=pd.DataFrame(user_input)
kclustering=kmeans.predict(user_df)
print("This belong to group:",kclustering)
```
